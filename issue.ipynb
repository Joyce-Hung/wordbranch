```python
#時間列表歷史紀錄(csv格式)
times=pd.read_csv("app_actions_2023_02_11.csv")
#times['time dt']=times['time'].apply(datetime.datetime.fromtimestamp)
times['date']=times['time'].apply(to_date)
unprs=times['action']=='unlock-prs'
times_unprs=times[unprs]
times_unprs
```

```python
#刪除 user-200167
#pd.set_option('display.max_rows',None)
#pd.set_option('display.max_columns',None)
acc_user=times_unprs['user']!=200167
del_user=times_unprs[acc_user]
del_user
```

```python
#issue2 - unlock-prs點擊次數/天
import datetime
from utils import to_date
#簡化資料-取日期
del_user[['action','date']]
#times_unprs=times_unprs[['action','dt_only']]
```

```python
#產出 unlock-prs點擊次數/天 的列表-except 200167
timelist=del_user.groupby('date')['action'].apply(pd.Series.count).reset_index()
timelist
```

```python
#產出 unlock-prs點擊次數/天 的列表- 200167 only
singlelist=single_user.groupby('date')['action'].apply(pd.Series.count).reset_index()
singlelist=singlelist.iloc[1:]
singlelist
```

```python
#issue1 - 使用unlock-prs的 unique user /天
unique_user=del_user[['user','action','date']]
#people['user'].unique()
user_countlist=unique_user.groupby('date')['user'].apply(pd.Series.nunique).reset_index()
user_countlist['avg_hits']=round(timelist['action']/user_countlist['user'],2)
user_countlist
```

```python
# (CHART)unlock-prs 點擊次數/天 + 使用 unlock-prs 的 unique user /天
# Ref-dic---.ipynb
labels=list(timelist['date'])# y-axis unit-date
act_counts=list(timelist['action'])# bar1-action counts
single_counts=list(singlelist['action'])#bar2-action counts(200167 only)
user_counts=list(user_countlist['user'])# bar3-unique user counts
avg_counts=list(user_countlist['avg_hits'])# bar4-average hits


x=np.arange(len(labels))
width=0.2 # widths of bars

fig, ax = plt.subplots()
fig.set_figheight(len(labels))

#bar tags
rec1=ax.barh(x+(width*1.5), act_counts, width, label='hit')#(bar center loc., data ,bar's width, bar tag)
rec2=ax.barh(x+(width*0.5),single_counts,width,label='hit-200167')
rec3=ax.barh(x-(width*0.5), user_counts, width, label='unique users')
rec4=ax.barh(x-(width*1.5),avg_counts,width,label='average hits')

ax.set_title('Hits & Unique Users per day of unlock-prs')
ax.set_ylabel('date')# hits count & unique_user count
ax.set_yticks(x, labels)
ax.legend()

ax.bar_label(rec1, padding=3)
ax.bar_label(rec2, padding=3)
ax.bar_label(rec3, padding=3)
ax.bar_label(rec4, padding=3)

fig.tight_layout()

#show fig.
plt.show()
```

```python
#issue3 - 特定一人當月點多少天
unprs_check=del_user[['user','date']]
unprs_check
```

```python
#產生列表
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
checklist=unprs_check.groupby('user')['date'].apply(pd.Series.nunique).reset_index()
checklist.sort_values(['date'],axis=0,ascending=False,inplace=True)
checklist
```

```python
#date_count list
daylist=checklist.groupby('date')['user'].apply(pd.Series.count).reset_index()
daylist
```

```python
# (CHART) unlock-prs 使用天數/月 對應 the number of users
# Ref-dic---.ipynb
labels=list(daylist['date'])# y-axis unit-date
user_counts=list(daylist['user'])# bar1-action counts

x=np.arange(len(labels))
width=0.3 # widths of bars

fig, ax = plt.subplots()
fig.set_figheight(len(labels))

#bar tags
rec1=ax.bar(x, user_counts, width, label='user(s)')#(bar center loc., data ,bar's width, bar tag)
#rec2=ax.barh(x-width/2, user_counts, width, label='unique users')


ax.set_title('unique user counts corresponding to unlock-prs day counts(per month)')
ax.set_xlabel('day(s)')# x-axis unit
ax.set_ylabel('user(s)')#y-axis unit
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rec1, padding=0.1)


fig.tight_layout()

#show fig.
plt.show()
```
